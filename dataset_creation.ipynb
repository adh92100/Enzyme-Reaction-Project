```python
import pandas as pd
from google.colab import drive
import os
import csv

drive.mount('/content/drive')


```

```python
names = os.listdir("/content/drive/My Drive/training_dataset/val_dataset")

for j in range(0,len(names)):


# reads files in order they are listed in the directory
  df = pd.read_csv("/content/drive/My Drive/training_dataset/val_dataset/" + names[j], header=None)

  num = len(df)
# initialize each variable
  EC = [''] * num
  Enzyme = [''] * num
  Reaction = [''] * num
  Source = [''] * num

# loop to get all of each value in each file
  for i in range(0, num):
    EC[i] = df.at[i,1]
    Enzyme[i] = df.at[i,0][df.at[i,0].find("|")+1:df.at[i,0].find(">")]
    Reaction[i] = df.at[i,0][:df.at[i,0].find("|")] + df.at[i,0][df.at[i,0].find(">"):]
    Source[i] = df.at[i,2]

  data = [EC, Enzyme, Reaction, Source] # organizes data

  data = [[row[i] for row in data] for i in range(len(data[0]))] # inverts data

# writes the file
  with open('/content/drive/My Drive/val.csv', mode='a', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)
```
